```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
%%time
train = pd.read_csv('../../TPS_2021/input/tabular-playground-series-feb-2021/train.csv')
test = pd.read_csv('../../TPS_2021/input/tabular-playground-series-feb-2021/test.csv')
```

```python
for i in range(10):
    le = LabelEncoder()
    train[f'cat{i}'] = le.fit_transform(train[f'cat{i}'].values)
    test[f'cat{i}'] = le.transform(test[f'cat{i}'].values)
train.head()
```

```python
train.to_csv('../../TPS_2021/input/tabular-playground-series-feb-2021/train_le.csv', index=False)
test.to_csv('../../TPS_2021/input/tabular-playground-series-feb-2021/test_le.csv', index=False)
```
